```python
%cd /
from glob import glob
img_list = glob('/content/data/images/*.jpg')
print(len(img_list))
```

```python
train_img_list, val_img_list = train_test_split(img_list, test_size=0.1, random_state=42) # 왜 txt는 안 나누지?
print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/data/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list)+'\n')

with open('/content/data/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list)+'\n')
```

```python
shutil.copy('/content/drive/Othercomputers/MyComputer/yamls/data.yaml', '/content/data/data.yaml')
```

```python
with open ('/content/data/data.yaml', 'r') as f:
    data = yaml.full_load(f)

print(data)

data['train'] = '/content/data/train.txt'
data['val'] = '/content/data/val.txt'

with open('/content/data/data.yaml', 'w') as f:
    yaml.dump(data, f)

print(data)
```
